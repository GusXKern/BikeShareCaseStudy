```python
import pandas as pd
```

```python
q2_2019 = pd.read_csv("/content/Divvy_Trips_2019_Q2.csv")
q3_2019 = pd.read_csv("/content/Divvy_Trips_2019_Q3.csv")
q4_2019 = pd.read_csv("/content/Divvy_Trips_2019_Q4.csv")
q1_2020 = pd.read_csv("/content/Divvy_Trips_2020_Q1.csv")
```

```python
print("\nQ2 2019 Data Info:")
q2_2019.info()
print("\nQ3 2019 Data Info:")
q3_2019.info()
print("\nQ4 2020 Data Info:")
q4_2019.info()
print("\nQ1 2020 Data Info:")
q1_2020.info()
```

```python
all_2019_trips = pd.concat([q2_2019, q3_2019,q4_2019], ignore_index=True)


all_2019_trips['trip_id'] = all_2019_trips['trip_id'].astype(str)
all_2019_trips['bikeid'] = all_2019_trips['bikeid'].astype(str)

print(all_2019_trips.head(10))
```

```python
q1_2020_test = q1_2020.drop(columns=['start_lat', 'start_lng', 'end_lat', 'end_lng', 'birthyear', 'gender',
'tripduration'], errors='ignore')
q1_2020_test.head(10)
```

```python
print("Columns for combo 2019:\n", all_2019_trips.columns)
print("\nColumns for Q1 2020:\n", q1_2020_test.columns)
```

```python
print("\nCombined 2019 Data Info:")
all_2019_trips.info()
print("\nQ1 2020 Data Info:")
q1_2020.info()
```

```python
all_trips = pd.concat([all_2019_trips,q1_2020_test], ignore_index=True)
all_trips.sort_values(by='start_time').head(10)
```

```python
# Inspect the new table that has been created
print("\nList of column names:\n", all_trips.columns)
print("\nHow many rows are in data frame?", len(all_trips))
print("\nDimensions of the data frame:", all_trips.shape)
print("\nSee the first 6 rows of data frame:\n", all_trips.head())
print("\nSee list of columns and data types:\n")
all_trips.info()
# Statistical summary of data (mainly for numerics)
print("\nStatistical summary of data:\n", all_trips.describe())
```

```python
all_trips['usertype'].value_counts()
```

```python
all_trips['usertype'] = all_trips['usertype'].replace({
'Subscriber': 'member',
'Customer': 'casual'
})
```

```python
all_trips['usertype'].value_counts()
```

```python
# Convert both columns using 'mixed' format to handle different styles
all_trips['start_time'] = pd.to_datetime(all_trips['start_time'], format='mixed')
all_trips['end_time'] = pd.to_datetime(all_trips['end_time'], format='mixed')

# Double check the results
print(all_trips[['start_time', 'end_time']].head())
```

```python
#Code to use start/end time to make date, month, day, year, day of the week columns
all_trips['date'] = all_trips['start_time'].dt.date
all_trips['month'] = all_trips['start_time'].dt.month
all_trips['day'] = all_trips['start_time'].dt.day
all_trips['year'] = all_trips['start_time'].dt.year
all_trips['day_of_week'] = all_trips['start_time'].dt.day_name()
```

```python
#Code to make ride_length column
all_trips['ride_length'] = (all_trips['end_time']-all_trips['start_time']).dt.total_seconds()
print("\nData types after adding date columns and ride_length:\n")
all_trips.info()
```

```python
#Using Boolean Masking to see how many negative values there are
all_trips['ride_length'].sort_values(ignore_index=True)

neg_ride = all_trips['ride_length'] < 0

len(all_trips[neg_ride])
```

```python
# Code to make a copy of the dataset with any rows with negative trip durations values excluded
all_trips_v2 = all_trips[(all_trips['from_station_name'] != "HQ QR") & (all_trips['ride_length'] >=
0)].copy()
```

```python
# Descriptive analysis on ride_length (all figures in seconds)
print("\nDescriptive statistics for ride_length:\n", all_trips_v2['ride_length'].describe())
```

```python
# Compare members and casual users
print("\nRide length statistics grouped by member_casual:\n",
all_trips_v2.groupby('usertype')['ride_length'].agg(['mean', 'median', 'max', 'min']))
```

```python
# See the average ride time by each day for members vs casual users
print("\nAverage ride length by member type and day of the week:\n",
all_trips_v2.groupby(['usertype', 'day_of_week'])['ride_length'].mean())
```

```python
# The days of the week need to be ordered for correct sorting for analysis
days_order = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
all_trips_v2['day_of_week'] = pd.Categorical(all_trips_v2['day_of_week'],
categories=days_order, ordered=True)
```

```python
# Now, run the aggregation again to see the sorted result
print("\nAverage ride length (sorted by day of week):\n", all_trips_v2.groupby(['usertype',
'day_of_week'])['ride_length'].mean())
```

```python
# Analyze ridership data by type and weekday
summary_stats = all_trips_v2.groupby(['usertype', 'day_of_week']).agg(
number_of_rides=('trip_id', 'count'),
average_duration=('ride_length', 'mean')
).reset_index()
```

```python
print("\nSummary of rides and duration by rider type and weekday:\n", summary_stats)
```

```python
counts_2 = all_trips_v2.groupby(['usertype', 'month']).agg(
number_of_rides=('trip_id', 'count'),
average_duration=('ride_length', 'mean')
)

print(counts_2)
# Reset the index so 'usertype' and 'month' become regular columns
counts_2_final = counts_2.reset_index()

# Now export the new version
counts_2_final.to_csv('avg_ride_length_month_2.csv', index=False)

from google.colab import files
files.download('avg_ride_length_month_2.csv')
```

```python
# 1. Create the lookup table from your 2020 file
station_lookup = q1_2020[['to_station_id', 'end_lat', 'end_lng']].drop_duplicates(subset=['to_station_id'])

# 2. Add 'to_station_id' to the groupby so it's preserved
grouped = all_trips_v2.groupby(['usertype', 'to_station_name', 'to_station_id']).agg(
    number_of_rides=('trip_id', 'count')
)

# 3. Get top 5 and use reset_index() to make 'to_station_id' a column again
top_5_per_user = grouped.sort_values(['usertype', 'number_of_rides'], ascending=[True, False]) \
                        .groupby('usertype') \
                        .head(5) \
                        .reset_index()  # <--- THIS IS THE KEY FIX

# 4. Now the merge will work because 'to_station_id' exists as a column
top_5_mapped = top_5_per_user.merge(station_lookup,
                                    on='to_station_id',
                                    how='left')

print(top_5_mapped)

```

```python
top_5_mapped.to_csv('top_5_lat.csv', index=False)
from google.colab import files
files.download('top_5_lat.csv')
```
